```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2-Create a Data frame

```python
dataframe=pd.read_csv('Zomato data .csv')
print(dataframe)
```

```python
dataframe
```

Convert the data type of column-rate

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

Type of Restaurant

```python
dataframe.head()
```

```python
dataframe.tail()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

Conclusion-Majority of the restaurant falls in dinning category

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker='o')
plt.xlabel('Types of restaurant',c="red",size=15)
plt.ylabel('Total votes',c="red",size=15)
```

Conclusion-Dinning restaurants has received maximum votes.

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()
```

Conclusion- The majority restaurants received ratings from 3.5 to 4.10

Average order spending by couples

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

Conclusion- the majority of couples preferr restaurants with an approximate cost of 300 rupees.

Which mode receives maximum rating

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

Conclusion- offline order received lower rating in comparison to online order.

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Online order")
plt.ylabel("Listed In (Type)")
plt.show()
```

Conclusion- Dining restaurants primarily accepr offline orders,whereas cafes primarily receive online orders. This suggests that clients preferred orders in person at restaurants,but prefer online ordering at cafes.
